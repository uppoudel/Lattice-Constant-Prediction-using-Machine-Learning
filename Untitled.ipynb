## General Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, r2_score
```

## Algorithm Imports

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor

# Linear, ElasticNet and Lasso Regression
from sklearn.linear_model import LinearRegression, MultiTaskLassoCV, MultiTaskElasticNetCV

#Support Vector Machine
from sklearn.svm import SVR, LinearSVR

#Decision Tree
from sklearn.tree import DecisionTreeRegressor

#Ensembes Models -> AdaBoost, GradientBoost, RandomForest, VotingEnsembleModel 
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor, RandomForestRegressor, VotingRegressor
```

```python
#PAD Formula Definition
pad = lambda y_true, y_pred: ( np.abs(y_true - y_pred) / y_true) * 100
```

## Data Import

```python
columns = ["sn", "compound", "ra", "rb", "xa", "xb", "za", "a", "b", "c"]
inputs = {0:"ra", 1:"rb", 2:"xa", 3:"xb", 4:"za"}
mapping = {0:"a", 1:"b", 2:"c"}
```

```python
data = pd.read_excel("Lattice Constant - Data Collection.xlsx", header=None)
data.columns = columns
```

```python
data = data.iloc[2:,]
```

## Seperating Dependent and Independent Variables

```python
X = data.iloc[:, [2,3,4,5,6]].dropna()
y = data.iloc[:, 7:].dropna()
```

```python
r2_scores = {}
predictions = {}
```

## K-Nearest Neighbour (KNN)

```python
#KNN
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X, y)
y_pred = knn.predict(X)
r2_scores["KNN"] = round(r2_score(y, y_pred), 3)
predictions["KNN"] = y_pred
print(f"R2 Score: {round(r2_score(y, y_pred), 3)}")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 4))

for i in range(3):
    ax[i].scatter(y.values[:,i], y_pred[:,i], c='w', edgecolors='b')
    ax[i].plot(y.values[:,i],y.values[:,i], c='r')
    ax[i].set_xlabel(f"Experimental {mapping[i]}")
    ax[i].set_ylabel(f"Predicted {mapping[i]}")

fig.suptitle(f'KNN with N=3')
plt.show()
```

## Linear Regression

```python
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred = linear_model.predict(X)
print(f"R2 Score: {round(r2_score(y, y_pred), 3)}")
r2_scores["Linear\nRegression"] = round(r2_score(y, y_pred), 3)
predictions["Linear Regression"] = y_pred
```

```python
coffs = np.round(linear_model.coef_, 3)
intercepts = np.round(linear_model.intercept_, 3)
```

```python
print("Approximated Equations from Linear Model:")
for i in mapping:
    print(f"{mapping[i]} = {coffs[i][0]}{inputs[0]} + {coffs[i][1]}{inputs[1]} + {coffs[i][2]}{inputs[2]} + {coffs[i][3]}{inputs[3]} + {coffs[i][4]}{inputs[4]} + {intercepts[i]}")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 4))

for i in range(3):
    ax[i].scatter(y.values[:,i], y_pred[:,i], c='w', edgecolors='b')
    ax[i].plot(y.values[:,i],y.values[:,i], c='r')
    ax[i].set_xlabel(f"Experimental {mapping[i]}")
    ax[i].set_ylabel(f"Predicted {mapping[i]}")

fig.suptitle(f'Linear Regression Model')
plt.show()
```

## Support Vector Machine

##### Linear SVM

```python
y_pred = []
r2s = []
coffs = []
intercepts = []
for i in mapping:
    linear_svm = LinearSVR()
    linear_svm.fit(X, y.iloc[:, i])
    y_pred.append(linear_svm.predict(X))
    r2s.append(round(r2_score(y.iloc[:, i], linear_svm.predict(X)), 3))
    coffs.append( np.round(linear_svm.coef_, 3))
    intercepts.append(np.round(linear_svm.intercept_[0], 3))
y_pred = np.array(y_pred).T
print(f"R2 Score: {round(np.array(r2s).mean(), 3)}")
r2_scores["Linear SVM"] = round(np.array(r2s).mean(), 3)
predictions["Linear SVM"] = y_pred
```

```python
print("Approximated Equations from Linear SVM Model: ")
for i in mapping:
    print(f"{mapping[i]} = {coffs[i][0]}{inputs[0]} + {coffs[i][1]}{inputs[1]} + {coffs[i][2]}{inputs[2]} + {coffs[i][3]}{inputs[3]} + {coffs[i][4]}{inputs[4]} + {intercepts[i]}")
```

#### Degree 2 Polynomial SVM

```python
y_pred = []
r2s = []
coffs = []
intercepts = []
for i in mapping:
    svm = SVR(kernel='poly', degree=2)
    svm.fit(X, y.iloc[:, i])
    y_pred.append(svm.predict(X))
    r2s.append(round(r2_score(y.iloc[:, i], svm.predict(X)), 3))
    coffs.append( np.round(svm.dual_coef_ , 3))
    intercepts.append(np.round(svm.intercept_[0], 3))
y_pred = np.array(y_pred).T
print(f"R2 Score: {round(np.array(r2s).mean(), 3)}")
r2_scores["Polynomial\nSVM (Degree 2)"] = round(np.array(r2s).mean(), 3)
predictions["Polynomial SVM (Degree 2)"] = y_pred
```

### Adaptive Boosting Model

```python
y_pred = []
r2s = []
feature_imps = []
for i in mapping:
    ada = AdaBoostRegressor(DecisionTreeRegressor(max_depth=1), n_estimators=100) 
    ada.fit(X, y.iloc[:,i])
    y_pred.append(ada.predict(X))
    r2s.append(round(r2_score(y.iloc[:, i], ada.predict(X)), 3))
    feature_imps.append(ada.feature_importances_)

y_pred = np.array(y_pred).T
print(f"R2 Score: {round(np.array(r2s).mean(), 3)}")
r2_scores["Adaptive\nBoosting\nRegression"] = round(np.array(r2s).mean(), 3)
predictions["Adaptive Boosting Regression"] = y_pred
```

```python
for feature in mapping:
    print(f"Feature Importance for {mapping[feature]}:")
    for i,c in enumerate(columns[2:-3]):
        print(f"{c}:\t{feature_imps[feature][i].round(3)}")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 4))

for i in range(3):
    ax[i].scatter(y.values[:,i], y_pred[:,i], c='w', edgecolors='b')
    ax[i].plot(y.values[:,i],y.values[:,i], c='r')
    ax[i].set_xlabel(f"Experimental {mapping[i]}")
    ax[i].set_ylabel(f"Predicted {mapping[i]}")

fig.suptitle(f'Adaptive Boosting Model (100 Trees)')
plt.show()
```
